```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

### (1) Define decay time experiment

We know that the decay time has an exponential distribution.<br>
Use NumPy library to generate random numbers from this distribution

```python
# Function for generating decay time for given average lifetime

def DecayTime(tmean):
    y = np.random.exponential(tmean)
    return y
```

### (2) Run multiple experiments

```python
# Number of decays to measure

Ntry = 100

# Mean lifetime - use muon as an example of unstable particle 

tmean = 2.2   # µs

# Array to store results

exp_data = np.empty(Ntry)

# Main experiment loop (more general concept)

for i in  range(Ntry):
    exp_data[i] = DecayTime(tmean)
    
# Please note that for our simple case one could generate all values in one call:
#
# exp_data = np.random.exponential(tmean,size=Ntry)
```

### (3) Plot results

```python
# Hiostogram range

tmax = 20.  # maximum value [µs]
Nbin = 20   # number of bins

# Plot histogram

plt.figure(figsize=(9, 6))

plt.hist(exp_data,bins=Nbin, range=[0.,tmax], density=False, log=False, histtype='step',linewidth=3)

plt.xlabel('Decay time [µs]',size=24)
plt.ylabel('Count',size=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title("Measured decay time",size=24)
plt.xlim([0.,tmax])

fname = '03_decay_'+str(Ntry)+'.png'
plt.savefig(fname)

plt.show()
```

### (4) Plot event fraction

```python
# Hiostogram range

tmax = 20.  # maximum value [µs]
Nbin = 20   # number of bins

# Event fraction is obtained by normalizing the histogram to 1 (density=True)

plt.figure(figsize=(9, 6))

plt.hist(exp_data,bins=Nbin, range=[0.,tmax], density=True, log=False, histtype='step',linewidth=3)

plt.xlabel('Decay time [µs]',size=24)
plt.ylabel('Event fraction',size=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title("Measured decay time",size=24)
plt.xlim([0.,tmax])

# Expected exponential distribution

def myExp(x):
    return np.exp(-x/tmean)/tmean

tref = np.linspace(0.,tmax,100)
pref = myExp(tref)

plt.plot(tref,pref,'--',linewidth=2)


fname = '03_decay2_'+str(Ntry)+'.png'
plt.savefig(fname)

plt.show()
```
