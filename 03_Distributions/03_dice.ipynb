```python
import numpy as np
import matplotlib.pyplot as plt
```

### (1) Define dice roll experiment

```python
# Returns random integer number from 1 to 6

def DiceRoll():
    y = np.random.randint(1,7)   # Range definition as in numpy arrays: including begining, excluding end value !
    return y
```

### (2) Run multiple experiments

```python
# Number of dice rolls

Ntry = 100

# Array to store results (count numbers), initialized to zeros

exp_data = np.zeros(6, dtype=int)

# Main experiment loop

for i in  range(Ntry):
    val = DiceRoll()
    exp_data[val-1] += 1     # Array inexes are from 0 to 5
```

### (3) Plot results

```python
# Plot count vs number

dice = np.arange(1,7,dtype=int)

plt.figure(figsize=(9, 6))

plt.plot(dice,exp_data,'o',markersize=10)
plt.xlabel('Result',size=24)
plt.ylabel('Count',size=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlim([0.5,6.5])
plt.ylim([0.,0.25*Ntry])      # Comment for close-up view (for large Ntry)
plt.title("Dice roll experiment",size=24)

fname = '03_dice_'+str(Ntry)+'.png'
plt.savefig(fname)

plt.show()
```

### (4) Plot probability

```python
# Plot probability vs number

prob_data = exp_data/Ntry     # just normalize to the number of experiments

plt.figure(figsize=(9, 6))

plt.plot(dice,prob_data,'o',markersize=10)
plt.xlabel('Result',size=24)
plt.ylabel('Probability',size=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlim([0.5,6.5])
plt.ylim([0.,0.25])      # Comment for close-up view (for large Ntry)
plt.title("Dice roll experiment",size=24)

# Reference line

xref = np.array([0.5,6.5])
yref = np.ones(2)/6.
plt.plot(xref,yref,'--',linewidth=2)

fname = '03_dice2_'+str(Ntry)+'.png'
plt.savefig(fname)

plt.show()
```
